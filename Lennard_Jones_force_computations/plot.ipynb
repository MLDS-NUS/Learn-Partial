```python
zc = "#4865A9"
xc =  "#EF8A43"
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme(style="ticks")

fig = plt.figure(figsize=(8,6),dpi=300)
ax = fig.add_subplot(1, 1, 1)
plt.yscale('log')
atoms = [800, 2700, 6400, 21600]
N_lz= [1500*800, 750*2700, 750*6400, 500*21600]
N_lx= [2250*50, 1750*50, 2000*50, 1250*50]

ax.plot(atoms, N_lz, 'o-', label='$\mathcal{L}_\mathbf{z}$',c=zc,linewidth=4, markersize=10)
ax.plot(atoms, N_lx, 'o-', label='$\mathcal{L}_{\mathbf{x},p}$',c=xc,linewidth=4, markersize=10)

ax.set_xlabel('# of atoms',fontsize=24)
ax.set_ylabel('# of force computations',fontsize=24)
ax.legend(fontsize=28, loc='best')
ax.set_ylim(5e4, 2e7)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
sns.despine()
plt.savefig('img/forces.pdf')
plt.show()
```
